```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
url_link = "https://www.squareyards.com/resale/search?cityId=0&propertyTypeName=&buildingType=2&"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}

driver.get(url_link)


def scroll_to_bottom(driver):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait for some time for content to load
    time.sleep(3)


def extract_from_site(soup, data):

    listings = soup.find_all("div", class_="tlWrp")
    # listings = soup.find_all("div", class_="dseprojectdata")

    print(listings)
    for prop in listings:

        script_tag = prop.find_all("script", type="application/ld+json")

        for scr in script_tag:
            script_content = scr.string.strip()
            data.append(json.loads(script_content))

    return data


i = 0
while True:
    page_source = driver.page_source

    scroll_to_bottom(driver)

    # Get the updated page source
    new_page_source = driver.page_source

    if page_source == new_page_source:
        break

    if i == 50:
        break
    i += 1
```

```python
soup = BeautifulSoup(page_source, "lxml")

# Extract the currently loaded listings
data = []
data = extract_from_site(soup, data)

driver.quit()
df = pd.DataFrame(data, index=None)

df.drop(
    columns=[
        "@type",
        "numberOfRooms",
        "numberOfBathroomsTotal",
        "floorlevel",
        "videoQuality",
        "contentSize",
        "duration",
        "embedUrl",
        "productionCompany",
        "uploadDate",
        "contentUrl",
        "thumbnailUrl",
    ],
    axis=1,
    inplace=True,
)


print(df.head())
df.to_csv("data2.csv")
```
